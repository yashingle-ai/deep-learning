```python
import numpy as np

docs = ['go india',
		'india india',
		'hip hip hurray',
		'jeetega bhai jeetega india jeetega',
		'bharat mata ki jai',
		'kohli kohli',
		'sachin sachin',
		'dhoni dhoni',
		'modi ji ki jai',
		'inquilab zindabad']
```

```python
from collections import defaultdict

word_index = defaultdict(lambda: word_index["<nothing>"])
word_index["<nothing>"] = 0

idx = 1
for text in docs:
    for word in text.split():
        if word not in word_index:
            word_index[word] = idx
            idx += 1
```

```python
sequences = [
    [word_index[word] for word in text.split()]
    for text in docs
]
```

```python
from torch.nn.utils.rnn import pad_sequence
import torch

# convert lists to tensors
sequences = [torch.tensor(seq) for seq in sequences]

# pad sequences
padded_sequences = pad_sequence(
    sequences,
    batch_first=True,
    padding_value=0   # <nothing> token
)
```
